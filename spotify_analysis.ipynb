```python
import glob

files = glob.glob("data/*.json")

files
```

```python
import pandas as pd
import json

dfs = []

for file in files:
    with open(file, "r", encoding="utf-8") as f:
        data = json.load(f)
        dfs.append(pd.DataFrame(data))

df = pd.concat(dfs, ignore_index=True)

df.head()
```

```python
df.columns
```

```python
# Convert timestamp to datetime
df["ts"] = pd.to_datetime(df["ts"])

# Rename long Spotify column names to simpler names
df = df.rename(columns={
    "master_metadata_track_name": "track",
    "master_metadata_album_artist_name": "artist",
    "master_metadata_album_album_name": "album"
})

# Keep only rows that are actual songs
df_music = df.dropna(subset=["track", "artist"]).copy()

# Convert milliseconds listened to minutes
df_music["minutes_played"] = df_music["ms_played"] / 60000

# Create useful time columns
df_music["date"] = df_music["ts"].dt.date
df_music["year"] = df_music["ts"].dt.year
df_music["month"] = df_music["ts"].dt.to_period("M")
df_music["day_of_week"] = df_music["ts"].dt.day_name()
df_music["hour"] = df_music["ts"].dt.hour

df_music.head()
```

```python
df_music.shape
```

```python
df_music[["track", "artist", "album", "minutes_played", "year", "month", "day_of_week", "hour"]].head()
```

```python
top_artists = (
    df_music.groupby("artist")["minutes_played"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

top_artists
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
top_artists.plot(kind="bar")
plt.title("Top 10 Artists by Minutes Played")
plt.xlabel("Artist")
plt.ylabel("Minutes Played")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
top_songs = (
    df_music.groupby(["track", "artist"])["minutes_played"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

top_songs
```

```python
monthly_listening = (
    df_music.groupby("month")["minutes_played"]
    .sum()
)

monthly_listening.tail()
```

```python
plt.figure(figsize=(12, 6))
monthly_listening.plot(kind="line", marker="o")
plt.title("Spotify Listening Over Time")
plt.xlabel("Month")
plt.ylabel("Minutes Played")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
